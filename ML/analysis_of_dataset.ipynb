```
import pandas as pd
import numpy as np
import seaborn as sns
import math
from matplotlib import pyplot as plt
```

```
covid_data=pd.read_csv("covid19india.csv")
```

```
sns.displot(covid_data['status'])
```

```
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 1})
sns.displot(covid_data['reportedOn'])
```

```
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 10})
# Create figure and plot space
fig, ax = plt.subplots(figsize=(100,100))
# Add x-axis and y-axis
ax.bar( covid_data['reportedOn'],covid_data.index.values,
       color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="count",
       title="Daily Total for covid patient amount")

plt.show()
```
